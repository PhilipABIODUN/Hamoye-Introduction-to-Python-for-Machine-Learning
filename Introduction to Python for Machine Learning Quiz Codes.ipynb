```python
import pandas as pd #imports pandas
Fuel_Data1 = pd.DataFrame(pd.read_csv('https://raw.githubusercontent.com/WalePhenomenon/climate_change/master/fuel_ferc1.csv', error_bad_lines=False))
print(Fuel_Data1) #prints Fuel_Data1
Fuel_Data1.isnull() # Checks if there are missing values in Fuel_Data1
print(Fuel_Data1.columns[Fuel_Data1.isnull().any()]) # Prints the columns with missing values
Fuel_Data=Fuel_Data1.dropna() # drops the rows with missing values and assigns the new dataframe to Fuel_Data
print(Fuel_Data) #prints Fuel_Data
```

```python
Fuel_Data.describe() #gets a summary description of the dataset (Fuel_Data)
print(Fuel_Data['fuel_qty_burned'].skew()) #prints the skewness of fuel quantity burned
print(Fuel_Data['fuel_qty_burned'].kurt()) #prints the kurtosis of fuel quantity burned
```

```python
print(Fuel_Data.corr()) #prints the correlation matrix for the dataset (Fuel_Data)
```

```python
FCPUD1994 = Fuel_Data[Fuel_Data["report_year"]==[1994]] #This extracts all the entries for the year 1994 from the Fuel_Data
FCPUD1998 = Fuel_Data[Fuel_Data["report_year"]==[1998]] #This extracts all the entries for the year 1998 from the Fuel_Data
print(FCPUD1994)
print(FCPUD1998)
print(len(FCPUD1994))
print(len(FCPUD1998))
# Length of the two are not equal
start_p= FCPUD1994['fuel_cost_per_unit_burned'].mean() #This extracts and gets the mean value for fuel cost per unit burned for the year 1994 from the Fuel_Data and assigns it as the starting point (start_p)
curr_p=FCPUD1998['fuel_cost_per_unit_burned'].mean() #This extracts and gets the mean value for fuel cost per unit burned for the year 1998 from the Fuel_Data and assigns it as the current point (curr_p)
perc_t=((curr_p-start_p)/start_p)*100  #This calculates the percentage increase
print(perc_t)  #This prints the percentage increase.
```
